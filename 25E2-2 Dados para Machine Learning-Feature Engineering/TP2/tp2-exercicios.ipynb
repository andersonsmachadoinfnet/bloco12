```python
import pandas as pd
from sklearn.datasets import load_iris, load_breast_cancer
from sklearn.preprocessing import minmax_scale, StandardScaler, Normalizer 

iris = load_iris()
features = iris['data']
target = iris['target']
```

```python
#1.Utilize a normalização StandardScaler para escalonar as features do Iris Dataset
df = pd.DataFrame(iris.data, columns=iris.feature_names)
sepal_len = df[['sepal length (cm)']]
df['zscore'] = StandardScaler().fit_transform(sepal_len)
desvio_padrao = df['zscore'].std()
media = df['zscore'].median()
print(f"Média: {media}")
print(f"Desvio Padrão: {desvio_padrao}")
print (df['zscore'])
```

```python
#3.Realize a regularização das features do Iris Dataset utilizando a norma Euclidiana (L2)
features_l2_norm = Normalizer(norm="l2").transform(features)
print("L2 Normatização:")
print(features_l2_norm)
```

```python
#3.Realize a regularização das features do Iris Dataset utilizando a norma Manhattan (L1)
features_l1_norm = Normalizer(norm="l1").transform(features)
print("L1 Normatização:")
print(features_l1_norm)
```
